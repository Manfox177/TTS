```python
import time
import gradio as gr
from TTS.api import TTS
import torch
from gradio import Video  # Added import

device = "cuda" if torch.cuda.is_available() else "cpu"
# tts = TTS("tts_models/en/multi-dataset/tortoise-v2").to(device)
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)

def predict(prompt, audio_file_pth, agree):
    if agree:  # Ensure consistent indentation
        try:
            output_filename = f"output/output_{str(time.time()).replace('.', '_')}.wav"
            # Unique filenames

            # Init
            # tts.tts(
                # text="abcdefghijklmnopqrstuvwxyz",
                # speaker_wav="clone/Adam.mp3",
                # language="en"
            # )

            tts.tts_to_file(
                text=prompt,
                file_path=output_filename,
                speaker_wav=audio_file_pth,
                language="en",
            )
            return (
                gr.make_waveform(
                    audio=output_filename,
                ),
                output_filename,  # Return the filename for potential usage
            )
        except Exception as e:
            return gr.Warning(f"TTS Error: {e}")
    else:
        return gr.Warning("Please accept the Terms & Conditions!")


gr.Interface(
    fn=predict,
    inputs=[
        gr.Textbox(
            label="Text Prompt",
            info="...",
            value="hello!",
        ),
        gr.Audio(
            label="Reference Audio",
            type="filepath",
            value="examples/female.wav",
        ),
        gr.Checkbox(
            label="Agree",
            value=False,
            info="I agree to the terms of the Coqui Public Model License at https://coqui.ai/cpml",
        ),
    ],
    outputs=[
        gr.Video(label="Waveform Visual"),
        gr.Audio(label="Synthesised Audio"),
    ],

).queue().launch(debug=False)
```
